**Pandas** is a powerful Python library for data manipulation and analysis. It provides two primary data structures: **Series** and **DataFrame**.  
A **Series** is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, etc.). It is similar to a column in a spreadsheet or a database table.  
A **DataFrame** is a two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns). It is similar to an Excel spreadsheet or SQL table, allowing for easy data manipulation, filtering, and analysis. Pandas makes data handling efficient and intuitive.

```python
import pandas as pd
```

A **Pandas Series** is a one-dimensional, labeled array in Python capable of holding data of any type, such as integers, floats, strings, or Python objects. Each element in a Series has a unique label, known as its index, which allows for intuitive data selection and alignment. Series are similar to columns in a spreadsheet or a single column in a DataFrame. They support a wide range of operations, including mathematical computations, filtering, and aggregation. Series are highly flexible, making them ideal for handling and analyzing data efficiently in various data science and machine learning tasks using the Pandas library.

```python
# PANDA
import pandas as pd
data=[1,2,3,4,5,]
series= pd.Series(data)
print("Series \n", series)
print(type(series))
```

df.loc[0] in pandas is used to access the row in the DataFrame df with the index label 0. It returns all the data for that row as a pandas Series, where the keys are the column names and the values are the corresponding data from that row.



```python
df.loc[0]
```

df.iloc[0] in pandas is used to access the first row of the DataFrame df by its integer position (zero-based index). It returns that row as a pandas Series, where the keys are the column names and the values are the data from the first row.

```python
df.iloc[0]
```
