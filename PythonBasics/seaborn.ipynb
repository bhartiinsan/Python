Data Visualization With Seaborn

Seaborn is a Python visualization library on metaplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualization with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
### Basic plotting with Seaborn
import seaborn as sns

tips = sns.load_dataset('tips')
tips
```

```python
## Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot of total bill of size")

```

```python
### Categorical Plots
## Bar Plot

sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar plot total bill by day')
plt.show()
```

```python
## Box Plot
sns.boxplot(x="day", y='total_bill', data=tips)
```

```python
## Violin Plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
## Histograms
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
## KDE PLOT
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
## Pairplot
sns.pairplot(tips)
```

```python
## Heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
