Here are some key points about missing values and their types in Python statistics:

### Missing Values
- Missing values are gaps in your dataset where data is not recorded.
- They can bias results and affect statistical analysis if not handled properly.

### Types of Missing Data
**1. Missing Completely at Random (MCAR):**
- The missingness is unrelated to any data, observed or unobserved.
- Example: Data lost due to a random technical error.
- MCAR data can be ignored or imputed without introducing bias.

**2. Missing at Random (MAR):**
- The missingness is related to observed data, but not the missing data itself.
- Example: Income missing more often for younger people.
- MAR can be handled using models that account for observed variables.

**3. Missing Not at Random (MNAR):**
- The missingness depends on the unobserved data itself.
- Example: People with high income are less likely to report their income.
- MNAR is the hardest to handle and may require advanced modeling.

### How to Measure Missingness in Python
- Use `df.isnull().sum()` to count missing values per column.
- Use `df.isnull().mean()` for the proportion of missing values.
- Visualize missingness with `missingno` library.

### Key Pointers
- Always check for missing values before analysis.
- Decide on a strategy: remove, impute, or model missing data.
- Use statistical tests (e.g., Littleâ€™s MCAR test) to check for MCAR.
- Understand the type of missingness to choose the best handling method.

## Imputation Missing Values
### 1- Mean Value Imputation

```python
# This error occurs because you are trying to calculate the mean of the 'Cabin' column, which contains string values (like cabin numbers or letters). The .mean() function only works with numeric data.

# How to fix:

# If you want to fill missing values in 'Cabin', use the mode (most frequent value), not the mean.

# working code

mode_value = df['Cabin'].mode()[0]
df['Cabin_mode'] = df['Cabin'].fillna(mode_value)
df[['Cabin_mode', 'Cabin']]


# Explanation:

# .mode()[0] gets the most common value in the column.
# .fillna(mode_value) fills missing values with that value.
# Gotcha:
# Never use .mean() on non-numeric columns. Use .mode() for categorical or string columns.


```

### How Median Imputation Works in Python

- `df['Parch']` selects the 'Parch' column (number of parents/children aboard).
- `df['Parch'].median()` calculates the median value of the column, ignoring missing values.
- `df['Parch'].fillna(df['Parch'].median())` replaces missing values in 'Parch' with the median.

**How Median Is Calculated:**
- The median is the middle value when all numbers are sorted.
- If there is an even number of values, itâ€™s the average of the two middle values.
- Example: For values `[0, 0, 1, 2, 5]`, the median is `1`.

### Why Data Analysts Use Median Imputation

- Median is less affected by outliers than the mean.
- Itâ€™s useful when the data is skewed or has extreme values.
- Imputing missing values with the median keeps the central tendency of the data.
- This helps in further analysis like regression, classification, or summary statistics, ensuring missing values do not bias results.

**Summary:**  
Median imputation fills gaps with a typical value, making your dataset more robust for analysis and modeling.

Hereâ€™s a step-by-step explanation of each line and its usefulness for data analysis:

---

### 1. `df[df['Cabin'].notna()]['Cabin'].mode()[0]`

- **What it does:**  
  Finds the most frequent (mode) value in the `'Cabin'` column, considering only non-missing values.
- **How:**  
  - `df['Cabin'].notna()` creates a boolean mask for non-missing values.
  - `df[df['Cabin'].notna()]` filters rows where `'Cabin'` is not null.
  - `['Cabin']` selects the `'Cabin'` column.
  - `.mode()[0]` returns the most common value.
- **Why useful:**  
  The mode is a good choice for filling missing values in categorical columns.

---

### 2. `mode_value = df[df['Cabin'].notna()]['Cabin'].mode()[0]`

- **What it does:**  
  Stores the most frequent `'Cabin'` value in the variable `mode_value`.
- **Why useful:**  
  You can use this value to fill all missing `'Cabin'` entries, ensuring consistency.

---

### 3. `df['Cabin_mode'] = df['Cabin'].fillna(mode_value)`

- **What it does:**  
  Creates a new column `'Cabin_mode'` where missing values in `'Cabin'` are replaced with the mode.
- **Why useful:**  
  Imputing missing values makes the dataset complete, which is essential for machine learning and statistical analysis.

---

### 4. `df['Cabin_mode'].isnull().sum()`

- **What it does:**  
  Counts the number of missing values in the new `'Cabin_mode'` column.
- **Why useful:**  
  After imputation, this should be zero, confirming all missing values were filled.

---

### 5. `df['Cabin'].isnull().sum()`

- **What it does:**  
  Counts the number of missing values in the original `'Cabin'` column.
- **Why useful:**  
  Shows how many missing values were present before imputation.

---

## **Why is this useful for data analysts?**

- **Ensures completeness:** Imputation fills gaps, so models and analyses donâ€™t fail or get biased.
- **Mode imputation for categorical data:** Keeps the most common value, which is often a reasonable guess for missing entries.
- **Easy to automate:** These steps can be applied to large datasets with many missing values.
- **Improves model performance:** Many algorithms require no missing values, so this step is crucial in big data workflows.

**Summary:**  
These steps help analysts clean and prepare data, making it ready for further analysis, visualization, or machine learning. Mode imputation is a simple, effective way to handle missing categorical data.
